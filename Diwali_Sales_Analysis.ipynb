```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df,palette={'M': 'blue', 'F': 'pink'})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# Most of the buyers arev females as depicted from above graph
```

```python
df.columns
```

```python
sx=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in sx.containers:
    sx.bar_label(bars)
```

```python
#for every age group we see women are dominating in shopping
#also the age group doing more shopping is 26-35
```

```python
pal = {
    '26-35': 'lightblue',
    '0-17': 'lightgreen',
    '18-25': 'lightcoral',
    '51-55': 'gold',
    '46-50': 'lightpink',
    '36-45': 'lightgrey',
    '55+':'red'
}

sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,palette=pal)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(25,10))
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#NO of order wise ranking arranging of state
```

```python
sales_states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(25,10))
sns.barplot(x='State',y='Amount',data=sales_states)
```

```python
#Amount spent wise ranking of state
```

```python
#also we see that UP is on top in terms of order or amount spent.
```

```python
df.columns
```

```python
px=sns.countplot(x='Marital_Status',data=df)
plt.figure(figsize=(5,20))
for bars in px.containers:
    px.bar_label(bars)
```

```python
#From above graph we can conclude that most shopping is done by married person.
```

```python
sales_s=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(6,5))
sns.barplot(x='Marital_Status',y='Amount',data=sales_s,hue='Gender')
```

```python
#From above grah we see that among married person most shopping is done by Women and have high purchasing power.
```

```python
sm=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in sm.containers:
    sm.bar_label(bars)
```

```python
sales_o=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_o)
```

```python
#from the above graph we see that most of the buyers are working in IT sector, Healthcare and Aviation.
```

```python
df.columns
```

```python
sm=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,4)})
for bars in sm.containers:
    sm.bar_label(bars)
```

```python
# from above graph we can conclude that buyers are spending most on clothing,food and electronic products
```

```python
sales_c=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_c)
```

```python
#Max amont spent was on food category- the difference is huge between 2nd and 3rd catgegory.
```
